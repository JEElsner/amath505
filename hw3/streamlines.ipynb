```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u(x, y):
    return np.array([x**2/y, y**2/x])
```

```python
def f(x, x_0=1, y_0=2):
    return np.sqrt(1 / (1/x**2 - 1/x_0**2 + 1/y_0**2))
```

```python
fig, ax = plt.subplots()

x = np.linspace(0.1, 2,20)
y = np.linspace(1, 3,20)
xs, ys = np.meshgrid(x, y)

us = u(xs, ys)

ax.quiver(xs, ys, *us)
curve_x = np.linspace(0.1, 1.1, 200)
ax.plot(curve_x, f(curve_x))

plt.show()
```
