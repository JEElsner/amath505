```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u(x, y):
    return np.array([x**2/y, y**2/x])
```

```python
def f(x, x_0=1, y_0=2):
    C = 0.5 * y_0**2 - x_0**2
    return np.sqrt(x**2 / (1 - C*x**2))
```

```python
fig, ax = plt.subplots()

x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)
xs, ys = np.meshgrid(x, y)

us = u(xs, ys)

ax.quiver(xs, ys, *us)
ax.plot(x, f(x))

plt.show()
```
